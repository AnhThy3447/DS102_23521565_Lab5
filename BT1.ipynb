**Word Segmentation**

```python
import json
```

```python
def read_sentences(name):
    data = json.load(open(f"/workspaces/DS102_23521565_Lab5/UIT-VSFC-{name}.json"))
    sentences = [item['sentence'] for item in data]
    return sentences
```

```python
sentences = []
names = ['train', 'dev', 'test']
for name in names:
    sentence = read_sentences(name)
    print(f"{name}:", len(sentence))
    sentences.extend(sentence)
print("Number of sentences: ", len(sentences))
```

```python
def word_segmentation(sentences):
    segmented_sentences = []
    for sentence in sentences:
        segmented_sentence = model.word_segment(sentence)
        segmented_sentences.append(segmented_sentence)

    segmented_sentences[0:5]
    return segmented_sentences
```

```python
segmented_sentences = word_segmentation(sentences)
segmented_sentences[0:5]
```
